```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python


(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python


plt.figure(figsize=(10, 3))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')
plt.show()
```

```python

from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),   # Input layer: flatten 28x28 images
    layers.Dense(128, activation='relu'),   # Hidden layer with 128 neurons
    layers.Dropout(0.2),                    # Prevent overfitting
    layers.Dense(10, activation='softmax')  # Output layer for 10 digits (0–9)
])


model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("✅ Model created and compiled successfully!")
```

```python


history = model.fit(x_train, y_train, epochs=5, validation_split=0.1)
```

```python


test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\n✅ Test accuracy: {test_acc * 100:.2f}%")
```

```python


predictions = model.predict(x_test)
```

```python
def plot_prediction(i):
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f"Predicted: {predictions[i].argmax()}, Actual: {y_test[i]}")
    plt.axis('off')
    plt.show()

plot_prediction(0)
plot_prediction(1)
plot_prediction(2)
```
